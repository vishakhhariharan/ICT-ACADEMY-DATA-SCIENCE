```python
import matplotlib.pyplot as plt
import pandas as pd

# importing the dataset
dataset = pd.read_csv('mushrooms.csv')
from sklearn.preprocessing import OrdinalEncoder
ord_enc = OrdinalEncoder()
dataset["cap-shape"] = ord_enc.fit_transform(dataset[["cap-shape"]])
dataset["cap-surface"] = ord_enc.fit_transform(dataset[["cap-surface"]])
dataset["cap-color"] = ord_enc.fit_transform(dataset[["cap-color"]])
dataset["bruises"] = ord_enc.fit_transform(dataset[["cap-shape"]])
dataset["odor"] = ord_enc.fit_transform(dataset[["odor"]])
dataset["gill-attachment"] = ord_enc.fit_transform(dataset[["gill-attachment"]])
dataset["gill-spacing"] = ord_enc.fit_transform(dataset[["gill-spacing"]])
dataset["gill-size"] = ord_enc.fit_transform(dataset[["gill-size"]])
dataset["gill-color"] = ord_enc.fit_transform(dataset[["gill-color"]])
dataset["stalk-root"] = ord_enc.fit_transform(dataset[["stalk-root"]])
dataset["stalk-shape"] = ord_enc.fit_transform(dataset[["stalk-shape"]])
dataset["stalk-color-above-ring"] = ord_enc.fit_transform(dataset[["stalk-color-above-ring"]])
dataset["stalk-color-below-ring"] = ord_enc.fit_transform(dataset[["stalk-color-below-ring"]])
dataset["stalk-surface-above-ring"] = ord_enc.fit_transform(dataset[["stalk-surface-above-ring"]])
dataset["stalk-surface-below-ring"] = ord_enc.fit_transform(dataset[["stalk-surface-below-ring"]])
dataset["veil-color"] = ord_enc.fit_transform(dataset[["veil-color"]])
dataset["veil-type"] = ord_enc.fit_transform(dataset[["veil-type"]])
dataset["ring-number"] = ord_enc.fit_transform(dataset[["ring-number"]])
dataset["ring-type"] = ord_enc.fit_transform(dataset[["ring-type"]])
dataset["spore-print-color"] = ord_enc.fit_transform(dataset[["spore-print-color"]])
dataset["population"] = ord_enc.fit_transform(dataset[["population"]])
dataset["habitat"] = ord_enc.fit_transform(dataset[["habitat"]])
# split the data into inputs and outputs
X = dataset.iloc[:, 1:22].values
y = dataset.iloc[:, 0].values

# training and testing data
from sklearn.model_selection import train_test_split

# assign test data size 25%
X_train, X_test, y_train, y_test =train_test_split(X, y, test_size=0.25, random_state=0)

```

```python
from sklearn.naive_bayes import BernoulliNB

# initializaing the NB
classifer=BernoulliNB()

# training the model
classifer.fit(X_train, y_train)

from sklearn import tree
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
y_pred2=clf.predict(X_test)
# testing the model
y_pred = classifer.predict(X_test)

# importing accuracy score
from sklearn.metrics import accuracy_score

# printing the accuracy of the model
print('nave bayes=',accuracy_score(y_test, y_pred))
print('decision tree=',accuracy_score(y_test, y_pred2))
```
