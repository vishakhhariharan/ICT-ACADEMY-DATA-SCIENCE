```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('Telco-Customer-Churn.xlsx')
df
```

```python
print("Values in each gender \n",df['gender'].value_counts())

plt.figure(figsize=(10,5))
sns.countplot(data = df,  x = 'gender', hue = 'Churn')
plt.title('Gender Vs Churn', fontsize=20,)
plt.xlabel('Gender')
plt.ylabel('Churn Count')
plt.show()
```

```python
df1= df[(df['gender']=='Female') & (df['SeniorCitizen']==1) ]
df1
```

```python
plt.figure(figsize=(10,5))

sns.catplot(x='gender',  col='SeniorCitizen', kind='count', data=df1) # Around 568 Female Senior Citizens
plt.title('No. of Female Senior Citizens', fontsize=20,)
plt.xlabel('Female Senior Citizens')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.scatterplot(x ='tenure', y ='TotalCharges',data = df)
plt.title('Tenure vs Total Charges',size = 20)
```

```python
# Finding No. 1
plt.figure(figsize=(13,8))

sns.countplot(x = 'PaperlessBilling', data = df, hue = "PaymentMethod")
plt.title('Payment methods Vs Paperless Billing',size=15)
plt.xlabel('Paperless Billing')
plt.ylabel('Count')
plt.show()
```

```python
# Finding No. 2
plt.figure(figsize=(13,8))

sns.countplot(x = 'gender', data = df, hue = "PaymentMethod")
plt.title('Payment methods based on Gender',size=15)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```
